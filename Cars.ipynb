```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv("C:/data analyst/tableau/bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize = (9, 6))
plt.bar(x = df_used_cars["Brand"],
       height = df_used_cars["Cars Listings"],
       color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 13) 
plt.yticks(fontsize = 13) 
plt.title("Cars Listings by Brend", fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize = 13 ) 
plt.savefig("Used Cars Bar2.png")
plt.show()
```
